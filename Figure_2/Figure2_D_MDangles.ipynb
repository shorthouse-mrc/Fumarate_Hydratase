```python
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import numpy as np
```

```python
matplotlib.rcParams['figure.figsize'] = [10, 6]
```

```python
with open ("../Data/Structure_data/MD_results/R233H_angle_331_233_484_reordered.xvg") as f:
    content = f.readlines()
```

```python
anglelist_R233H = []
for line in content:
    if line.startswith("#"):
        continue
    elif line.startswith("@"):
        continue
    else:
        time = line.split()[0]
        angle = float(line.split()[1])
        anglelist_R233H.append(angle)
```

```python
with open ("../Data/Structure_data/MD_results/WT_angle_331_233_484_reordered.xvg") as f:
    content = f.readlines()
```

```python
anglelist_WT = []
for line in content:
    if line.startswith("#"):
        continue
    elif line.startswith("@"):
        continue
    else:
        time = line.split()[0]
        angle = float(line.split()[1])
        anglelist_WT.append(angle)
```

```python
sns.set_style("ticks")
sns.distplot(anglelist_R233H, color = "#43A9DB")
sns.distplot(anglelist_WT, color = "#FB4459")
sns.despine()
plt.yticks([])
plt.xlabel("Angle (degrees)", size = 14)
plt.ylabel("Relative Frequency", size = 14)
plt.savefig("Figure2_D_MDangles.pdf")
```
