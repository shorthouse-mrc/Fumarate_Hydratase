```python
import pandas as pd
import numpy as np
```

```python
aaconv = {'CYS': 'C', 'ASP': 'D', 'SER': 'S', 'GLN': 'Q', 'LYS': 'K',
     'ILE': 'I', 'PRO': 'P', 'THR': 'T', 'PHE': 'F', 'ASN': 'N', 
     'GLY': 'G', 'HIS': 'H', 'LEU': 'L', 'ARG': 'R', 'TRP': 'W', 
     'ALA': 'A', 'VAL':'V', 'GLU': 'E', 'TYR': 'Y', 'MET': 'M',
     'H1S': 'H', 'H2S': 'H'}
```

```python
mutation_data = pd.read_excel("./FH_mutations_DB.xlsx")
```

```python
mutation_data
```

```python
mutation_data['A'], mutation_data['Mutation'] = mutation_data['Protein Change'].str.split('p.', 1).str

mutation_data = mutation_data[mutation_data["Mutation"] != "?"]
mutation_data = mutation_data[mutation_data["Mutation"] != "(=)"]
mutation_data["Mutation"] = mutation_data.apply(lambda x: x["Mutation"][1:-1], axis =1)
mutation_data = mutation_data[~mutation_data.Mutation.str.contains("\*")]
mutation_data = mutation_data[~mutation_data.Mutation.str.contains("dup")]
mutation_data = mutation_data[~mutation_data.Mutation.str.contains("del")]

mutation_data["AAfrom"] = mutation_data.apply(lambda x: x["Mutation"][:3], axis =1)
mutation_data["AAto"] = mutation_data.apply(lambda x: x["Mutation"][-3:], axis =1)
mutation_data["Residue"] = mutation_data.apply(lambda x: int(x["Mutation"][3:-3]), axis =1)
mutation_data = mutation_data[mutation_data["Residue"] >= 49]
mutation_data["AAfrom"] = mutation_data["AAfrom"].apply(lambda x: aaconv.get((x.upper())))
mutation_data["AAto"] = mutation_data["AAto"].apply(lambda x: aaconv.get((x.upper())))
mutation_data["Mutation"] = mutation_data.apply(lambda x: x["AAfrom"] + str(x["Residue"]) + x["AAto"], axis =1)
```

```python
mutation_data = mutation_data.drop(["A", "AAfrom", "AAto", "Residue"], axis =1)
```

```python
mutation_data
```

```python
# Categories are done manually on the resulting file
mutation_data.to_csv("./FH_mutations_DB_categories.csv")
```
