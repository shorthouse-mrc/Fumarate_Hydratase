```python
import matplotlib.pyplot as plt
import matplotlib
import glob
import os
import pandas as pd
import seaborn as sns
import re
from scipy import stats
import numpy as np
```

```python
matplotlib.rcParams['figure.figsize'] = [14, 8]
```

```python
# Load data frame containing all the mutation information
complete_data = pd.read_csv("../Data/Structure_data/FH_complete_mutation_summary.csv")
```

```python
f, ax = plt.subplots(figsize=(10, 10))
# Plot Rosetta vs Foldx
snsplot = sns.regplot(complete_data["Rosetta_energy"], complete_data["Foldx_energy"], line_kws={'color':'#FB4459'}, scatter_kws={'lw':'0', 'edgecolors':'white', 'alpha':'0.25', 'color':'#43A9DB'})
snsplot.set_xlim(auto=True)
ax.set_xlabel("Rosetta Predicted $\Delta \Delta$G (Kcal/mol)", size = 18)
ax.set_ylabel("Foldx Predicted $\Delta \Delta$G (Kcal/mol)", size = 18)

#symetric log for easier visualization of high energy
snsplot.set(xscale="symlog", yscale="symlog")
sns.despine()
plt.savefig("Figure_3_A_foldxvsrosettaenergy.png", dpi = 300)
```

```python
## Calculate the spearman rank
stats.spearmanr(complete_data["Rosetta_energy"], complete_data["Foldx_energy"])
```
